# 4 - Correlação
**Direto ao ponto**: correlação é o coeficiente de interdependência entre duas ou mais variáveis.

Em muitos casos, quando analisamos algumas variáveis, essa interdependência é muito clara. Exemplo:
- Massa corporal e altura,
- Consumo de combustível e distância percorrida,
- Número de anúncios e produtos vendidos,
- Tempo de estudo e desempenho na avaliação.

Entretanto, algumas vezes essa correlação pode não ser muito clara.

## Correlações positivas

No trecho abaixo são gerados alguns valores aleatórios para x e alguns valores aleatórios para y. Mas no caso o Y, foi feita uma operação somando ao array x, para mostrar a correlação positiva

```python
import numpy as np

np.random.seed(1)

# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Positive Correlation with some noise
y = x + np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

Como o array y é uma combinação do array X somado aos valores aleatórios, o gráfico irá mostrar que quando X aumenta, o Y aumenta também, denotando uma correlação positiva.

```python
# import matplotlib
import matplotlib.pyplot as plt

plt.scatter(x, y)
```

## Correlações negativas

Ocorre o contrário quando X cresce e Y descresce. A correlação será negativa como mostra o gráfico.

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Negative Correlation with some noise
y = 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

## Correlação fraca ou sem correlação

Já quando os valores são totalmente aleatórios, sem relação nenhuma, o gráfico mostra dados bagunçados

```python
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

## Correlação é diferente de causalidade!
