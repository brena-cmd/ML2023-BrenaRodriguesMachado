# 3 - Tipos e Conversão de tipos
O Numpy suporta uma variedade de tipos muito maior do que o Python.

Existem 5 tipos básicos numéricos
- Booleans (bool)
- Integers (int)
- Unsigned Intergers (uint)
- Floating Points (float)
- Complex

Complementarmente, a célula abaixo mostra um quadro comparativos com uma maior quantidade de tipos do Numpy.

```python
import numpy as np
```

Criando uma variável do tipo float32

```python
x = np.float32(1.0)
x
```

Criando um array de int

```python
y = np.int_([1,2,4])
y
```

A função arange cria um array que com valores determinados com start-stop-step. Quando os outros valores não são escecificados, o np criará um array indo de 0 até o valor determinado

```python
z = np.arange(3, dtype=np.uint8)
z
```

Criando array e determinando o tipo

```python
np.array([1, 2, 3], dtype='f')
```

## Convertendo tipos de arrays

Para convertermos arrays Numpy, basta utilizarmos o método astype().

É possíve converter um array já criado para outro tipo de dado utilizando astype

```python
z.astype(float)
```

```python
np.int8(z)
```

Para verificarmos o tipo de um array Numpy, basta verificarmos o seu atributo dtype.

```python
z.dtype
```
